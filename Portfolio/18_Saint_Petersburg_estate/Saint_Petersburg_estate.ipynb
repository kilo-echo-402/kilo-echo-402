```python
# Изучим статистику по колонке
df.living_area.describe()
```

Есть аномально небольшие значения, но разбираться с ними будем потом. Обработаем пропуски.

Хотелось бы заполнить пропуски разницей между площадью кухни и общей площадью. Но! В результате гугления выяснилось, что общая площадь не равна сумме площадей кухни и жилой зоны. Поэтому пойдём другим путём: для начала проверим, сильна ли корреляция между колонками `total_area` и `living_area`.

```python
# Взглянем на корреляцию между площадями.
df.total_area.corr(df.living_area)
```

Очень сильная положительная корреляция! Почти единица. Наверное можно говорить о полноценной зависимости между общей площадью и жилой площадью (подтвердили то, что казалось и так очевидным). Посмотрим на процентное отношение между двумя этими площадями.

```python
# Посчитаем отношение жилой площади к общей и взглянем на статистику
round(df['living_area'] / df['total_area'], 4).describe()
```

Медианное отношение жилой площади к общей - 0,5682  
Среднее не сильно отличается от медианы - 0,5649  
Думаю, что мы теперь запросто можем заполнить пропуски в `living_area` умножив общую площадь на медианное отношение площадей.

```python
df.living_area = df.living_area.fillna(df.total_area * 0.5682)
f'Количество пропущенных значений в колонке living_area: {df.living_area.isna().sum()}'
```

```python
# Ещё раз посчитаем статистику по отношению площадей для наглядности
round(df['living_area'] / df['total_area'], 4).describe()
```

Всё как и должно быть - медианное отношение площадей осталось прежним, а среднее немного изменилось.

### Этаж квартиры

```python
# Напишем функцию для определения категории этажа
def set_floor_category(row) -> str:
    """
    В зависимости от значения в столбце floor и с учетом столбца floors_total,
    необходимо указать категорию этажа: "первый", "последний" или "другой"
    """
    try:
        if row.floor == 1:
            return "первый"
        elif row.floor == row.floors_total:
            return "последний"
        else:
            return 'другой'
    except Error:
        print(f'Неккоректное значение: {row}')
```

```python
# Применим новую функцию к датасету с записью результатов в новый столбец
df['floor_category'] = df.apply(set_floor_category, axis=1)
df.head(5)
```
